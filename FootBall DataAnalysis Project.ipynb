# Create MinsPerMatch & GoalsPerMatch two new columns

```python
fb["MinsPerMatch"] = (fb["Mins"])/(fb["Matches"]).astype(int)
fb["GoalsPerMatch"] = (fb["Goals"])/(fb["Matches"]).astype(float)
```

# Pie Chart for penalties missed vs scored

```python
plt.figure(figsize=(16,9))
data=[(fb["Penalty_Attempted"].sum()-fb["Penalty_Goals"].sum()),(fb["Penalty_Goals"].sum())]
labels=["Penalties Missed","Penalties Scored"]
colors = sns.color_palette("Set2")
plt.pie(data,labels=labels,colors=colors,autopct= "%.0f%%")
plt.show()
```

# Players based on Age group

```python
under_20 = fb[fb.Age<=20]
age_20_to_25 = fb[(fb.Age>20)&(fb.Age<=25)]
age_25_to_30 = fb[(fb.Age>25)&(fb.Age<=30)]
above_30 = fb[fb.Age>=30]
```

```python
a = np.array([under_20.Name.count(),age_20_to_25.Name.count(),age_25_to_30.Name.count(),above_30.Name.count()])
mylabel = ["<=20","21_to_25","26_to_30",">=30"]
plt.title("Total players with age", fontsize = 20)
plt.pie(a,labels=mylabel,autopct="%.2f%%")
plt.show()
```

# Average age of players in each club

```python
plt.figure(figsize=(12,6))
plt.title("Average age of players in each club", fontsize=20)
sns.boxplot(data = fb, x= "Club", y= "Age")
plt.xticks(rotation=90)
plt.xlabel("Club",fontsize=20)
plt.ylabel("Age",fontsize=20)
plt.show()
```

```python
nplayers = fb.groupby("Club").size()
tage = fb.groupby("Club")["Age"].sum()
avg = tage / nplayers
avg.sort_values(ascending=False)
```

# Total assists from each club

```python
fb.head(1)
```

```python
plt.figure(figsize=(16,9))
nassists = pd.DataFrame(fb.groupby("Club",as_index=False)["Assists"].sum())
plt.title("Plot of Club vs Total Assists",fontsize=25)
sns.barplot(data = nassists , x = "Club", y="Assists")
plt.xticks(rotation = 90)
plt.xlabel("Club",fontsize=20)
plt.ylabel("Assists",fontsize=20)
plt.show()
```

# Top 10 Assists

# Goals by Club

```python
plt.figure(figsize=(16,9))
plt.title("Goals by Club", fontsize=25)
goal = pd.DataFrame(fb.groupby("Club",as_index=False).Goals.sum().sort_values(by="Goals",ascending=False))
sns.barplot(data=goal, x="Club",y="Goals")
plt.xticks(rotation=90)
plt.xlabel("Club", fontsize=20)
plt.ylabel("Goals",fontsize=20)
plt.show()
```

# Most Goals by player

```python
fb.sort_values(by="Goals",ascending=False).head(10)
```

```python
playergoal = fb[["Name","Club","Goals","Matches"]].nlargest(n=10,columns="Goals")
plt.figure(figsize=(16,9))
plt.title("Goals by Players",fontsize=25)
sns.barplot(data=playergoal,x="Name",y="Goals")
plt.xlabel("Players",fontsize=20)
plt.ylabel("Goals",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

# Goals With Assists and Without Assists

```python
plt.figure(figsize=(16,9))
g = fb.Goals.sum()
a = fb.Assists.sum()
d=[g-a,a]
label = ["Without Assists","With Assists"]
color = sns.color_palette("Set2")
plt.pie(d,labels=label,colors=color,autopct="%.2f%%")
plt.show()
```
